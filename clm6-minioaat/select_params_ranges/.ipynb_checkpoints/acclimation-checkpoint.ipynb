```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
# from ctsm
#ft(tl,ha) = np.exp( ha / (rgas*(tfrz+25)) * (1 - (tfrz+25)/tl) )
#fth(tl,hd,se,scaleFactor) = scaleFactor / ( 1 + np.exp( (-hd+se*tl) / (rgas*tl) ) )
#fth25(hd,se) = 1 + np.exp( (-hd+se*(tfrz+25)) / (rgas*(tfrz+25)) )
#fh = fth25/fth
```

```python
def f(Tv,ha):
    return np.exp((ha/(298.15*8.314))*(1-298.15/Tv))
```

```python
def fH(Tv,hd,dS):
    return (1+np.exp((298.15*dS-hd)/(298.15*8.314))) / (1+np.exp((dS*Tv-hd)/(8.314*Tv)))
```

```python
def acclimation_C3(Tv,T10,Patm,Vcmax25,Jmax25,
                   vcmaxha,vcmaxhd,vcmaxse_sf,
                   jmaxha,jmaxhd,jmaxse_sf,
                   tpuha,tpuhd,tpuse_sf,
                   lmrha,lmrhd,
                   kcha,koha,cpha):

    Tp25 = 0.167*Vcmax25
    Rd25 = 0.015*Vcmax25
    
    #  allow ‚àÜùëÜ to vary with growth temperature
    dS_vcmax = (668.39 - 1.07*(T10 - 273.15))*vcmaxse_sf #J/mol/K
    dS_jmax = (659.70 - 0.75*(T10 - 273.15))*jmaxse_sf #J/mol/K
    dS_tp = (668.39 - 1.07*(T10 - 273.15))*tpuse_sf #J/mol/K
    
    dS_rd = 490 #J/mol/K
    
    Vcmax = Vcmax25*f(Tv,vcmaxha)*fH(Tv,vcmaxhd,dS_vcmax)
    Jmax = Jmax25*f(Tv,jmaxha)*fH(Tv,jmaxhd,dS_jmax)
    Tp = Tp25*f(Tv,tpuha)*fH(Tv,tpuhd,dS_vcmax)
    Rd = Rd25*f(Tv,lmrha)*fH(Tv,lmrhd,dS_rd)

    #Kc,Ko,cp
    Kc25 = 404.9E-6*Patm # Pa
    Ko25 = 278.4E-3*Patm # Pa
    cp25 = 42.75E-6*Patm # Pa

    Kc = Kc25*f(Tv,kcha)
    Ko = Ko25*f(Tv,koha)
    cp = cp25*f(Tv,cpha)

    return Vcmax,Jmax,Tp,Kc,Ko,cp
```

### Test scale factors

```python
# parameters
# parameters
vcmaxha = 72000 
vcmaxhd = 200000 
vcmaxse_sf = 1
jmaxha = 50000 
jmaxhd = 200000 
jmaxse_sf = 1
tpuha = 72000
tpuhd = 200000
tpuse_sf = 1

Vcmax25 = 31.4  # ponderosa umol/gN/s
Jmax25 = 60.6   # ponderosa umol/gN/s
Tp25 = 0.167*Vcmax25

kcha = 79430
koha = 36380
cpha = 37830
```

```python
Tv = 273.15+25
T10 = 303
Patm = 101325 # Pa
Vcmax,Jmax,Tp,Kc,Ko,cp = acclimation_C3(Tv,T10,Patm,Vcmax25,Jmax25,
                   vcmaxha,vcmaxhd,vcmaxse_sf,
                   jmaxha,jmaxhd,jmaxse_sf,
                   tpuha,tpuhd,tpuse_sf,
                   lmrha,lmrhd,
                   kcha,koha,cpha)
```

```python
Tv = 310
Vcmax_sf = np.empty((9,20))
Jmax_sf = np.empty((9,20))
Tpu_sf = np.empty((9,20))

for t,T10 in enumerate(np.linspace(284,308,9)):

    for j,sf in enumerate(np.linspace(0.8,1.2,20)):
        Vcmax,Jmax,Tp,Kc,Ko,cp = acclimation_C3(Tv,T10,Patm,Vcmax25,Jmax25,
                       vcmaxha,vcmaxhd,sf,
                       jmaxha,jmaxhd,sf,
                       tpuha,tpuhd,sf,
                       lmrha,lmrhd,
                       kcha,koha,cpha)
        Vcmax_sf[t,j] = Vcmax
        Jmax_sf[t,j] = Jmax
        Tpu_sf[t,j] = Tp
```

```python
cmap = cm.get_cmap('viridis', Vcmax_sf.shape[1]) 
plt.figure(figsize=(10,4))
x = np.linspace(284,308,9)
for i in range(20):
    plt.subplot(1,3,1)
    plt.plot(x,Vcmax_sf[:,i],color=cmap(i))

    plt.subplot(1,3,2)
    plt.plot(x,Jmax_sf[:,i],color=cmap(i))

    plt.subplot(1,3,3)
    plt.plot(x,Tpu_sf[:,i],color=cmap(i))
```

### Test activation/deactivation energy parameters (Vcmax,Jmax,Tpu)

```python
# parameters
vcmaxha = 72000 
vcmaxha_range = np.linspace(45000,173000,10)
vcmaxhd = 200000 
vcmaxhd_range = np.linspace(90000,415000,10)
jmaxha = 50000 
jmaxha_range = np.linspace(34000,78000,10)
jmaxhd = 200000 
jmaxhd_range = np.linspace(88300,280000,10)
tpuha = 72000 
tpuha_range = np.linspace(45000,173000,10)
tpuhd = 200000 
tpuhd_range = np.linspace(90000,415000,10)

Vcmax25 = 31.4  # ponderosa umol/gN/s
Jmax25 = 60.6   # ponderosa umol/gN/s
Tpu25 = 0.167*Vcmax25
```

```python
# Activation Energy
Tv = np.linspace(285,308,20)
Vcmax= np.empty((20,10))
for i,ha in enumerate(vcmaxha_range):
    Vcmax[:,i] = Vcmax25*f(Tv,ha)*fH(Tv,vcmaxhd,490)

Jmax= np.empty((20,10))
for i,ha in enumerate(jmaxha_range):
    Jmax[:,i] = Jmax25*f(Tv,ha)*fH(Tv,jmaxhd,490)

Tpu= np.empty((20,10))
for i,ha in enumerate(tpuha_range):
    Tpu[:,i] = Tpu25*f(Tv,ha)*fH(Tv,tpuhd,490)
```

```python
plt.figure(figsize=(12,4))
cmap = cm.get_cmap('viridis', Vcmax.shape[1]) 
plt.subplot(1,3,1)
for i in range(10):
    plt.plot(Tv,Vcmax[:,i],color=cmap(i))
plt.plot(Tv,Vcmax25*f(Tv,vcmaxha)*fH(Tv,vcmaxhd,490),c='k',linestyle='--')
plt.ylabel('Vcmax')
plt.xlabel('Tveg (K)')
plt.title('Vcmax activation energy')

plt.subplot(1,3,2)
for i in range(10):
    plt.plot(Tv,Jmax[:,i],color=cmap(i))
plt.plot(Tv,Jmax25*f(Tv,jmaxha)*fH(Tv,jmaxhd,490),c='k',linestyle='--')
plt.ylabel('Jmax')
plt.xlabel('Tveg (K)')
plt.title('Jmax activation energy')

plt.subplot(1,3,3)
for i in range(10):
    plt.plot(Tv,Tpu[:,i],color=cmap(i))
plt.plot(Tv,Tpu25*f(Tv,tpuha)*fH(Tv,tpuhd,490),c='k',linestyle='--')
plt.ylabel('Tpu')
plt.xlabel('Tveg (K)')
plt.title('Tpu activation energy')

plt.tight_layout()
```

```python
# deactivation energy
Tv = np.linspace(285,308,20)
Vcmax= np.empty((20,10))
for i,hd in enumerate(vcmaxhd_range):
    Vcmax[:,i] = Vcmax25*f(Tv,vcmaxha)*fH(Tv,hd,490)

Jmax= np.empty((20,10))
for i,hd in enumerate(jmaxhd_range):
    Jmax[:,i] = Jmax25*f(Tv,jmaxha)*fH(Tv,hd,490)

Tpu= np.empty((20,10))
for i,hd in enumerate(tpuhd_range):
    Tpu[:,i] = Tpu25*f(Tv,tpuha)*fH(Tv,hd,490)
```

```python
plt.figure(figsize=(12,4))
cmap = cm.get_cmap('viridis', Vcmax.shape[1]) 
plt.subplot(1,3,1)
for i in range(10):
    plt.plot(Tv,Vcmax[:,i],color=cmap(i))
plt.plot(Tv,Vcmax25*f(Tv,vcmaxha)*fH(Tv,vcmaxhd,490),c='k',linestyle='--')
plt.ylabel('Vcmax')
plt.xlabel('Tveg (K)')
plt.title('Vcmax deactivation energy')

plt.subplot(1,3,2)
for i in range(10):
    plt.plot(Tv,Jmax[:,i],color=cmap(i))
plt.plot(Tv,Jmax25*f(Tv,jmaxha)*fH(Tv,jmaxhd,490),c='k',linestyle='--')
plt.ylabel('Jmax')
plt.xlabel('Tveg (K)')
plt.title('Jmax deactivation energy')

plt.subplot(1,3,3)
for i in range(10):
    plt.plot(Tv,Tpu[:,i],color=cmap(i))
plt.plot(Tv,Tpu25*f(Tv,tpuha)*fH(Tv,tpuhd,490),c='k',linestyle='--')
plt.ylabel('Tpu')
plt.xlabel('Tveg (K)')
plt.title('Tpu deactivation energy')

plt.tight_layout()
```

### Leaf maintenence respiration

```python
Vcmax25 = 31.4  # ponderosa umol/gN/s
Rd25 = 0.015*Vcmax25
lmrha = 46390
lmrha_range = np.linspace(0.5*lmrha,1.5*lmrha,10)
lmrhd = 150650
lmrhd_range = np.linspace(0.5*lmrhd,1.5*lmrhd,10)
lmrse = 490
lmrse_range = np.linspace(0.8*lmrse,1.2*lmrse,10)

dS = 490 #J/mol/K
```

```python
Tv = np.linspace(285,308,20)
Rd_ha= np.empty((20,10))
for i,ha in enumerate(lmrha_range):
    Rd_ha[:,i] = Rd25*f(Tv,ha)*fH(Tv,lmrhd,dS)

Rd_hd= np.empty((20,10))
for i,hd in enumerate(lmrhd_range):
    Rd_hd[:,i] = Rd25*f(Tv,lmrha)*fH(Tv,hd,dS)

Rd_se= np.empty((20,10))
for i,se in enumerate(lmrse_range):
    Rd_se[:,i] = Rd25*f(Tv,lmrha)*fH(Tv,lmrhd,se)
```

```python
plt.figure(figsize=(12,4))
cmap = cm.get_cmap('viridis', Vcmax.shape[1]) 
plt.subplot(1,3,1)
for i in range(10):
    plt.plot(Tv,Rd_ha[:,i],color=cmap(i))
plt.plot(Tv,Rd25*f(Tv,lmrha)*fH(Tv,lmrhd,490),c='k',linestyle='--')
plt.ylabel('Rd')
plt.xlabel('Tveg (K)')
plt.title('Rd activation energy')

plt.subplot(1,3,2)
for i in range(10):
    plt.plot(Tv,Rd_hd[:,i],color=cmap(i))
plt.plot(Tv,Rd25*f(Tv,lmrha)*fH(Tv,lmrhd,490),c='k',linestyle='--')
plt.ylabel('Rd')
plt.xlabel('Tveg (K)')
plt.title('Rd deactivation energy')

plt.subplot(1,3,3)
for i in range(10):
    plt.plot(Tv,Rd_se[:,i],color=cmap(i))
plt.plot(Tv,Rd25*f(Tv,lmrha)*fH(Tv,lmrhd,490),c='k',linestyle='--')
plt.ylabel('Rd')
plt.xlabel('Tveg (K)')
plt.title('Rd entropy')

plt.tight_layout()
```

### Kc, Ko, Cp 

```python
Patm = 101325 # Pa
Kc25 = 404.9E-6*Patm
Ko25 = 278.4E-3*Patm
cp25=42.75E-6*Patm

kcha = 79430
kcha_range = np.linspace(0.5*kcha,1.5*kcha,10)
koha = 36380
koha_range = np.linspace(0.5*koha,1.5*koha,10)
cpha = 37830
cpha_range = np.linspace(0.5*cpha,1.5*cpha,10)
```

```python
def ha_acclim(Tv,def25,ha):
    return def25*f(Tv,ha)
```

```python
# Kc
Tv = np.arange(285,313,1)
Kc = np.empty((len(Tv),10))
for i in range(10):
    Kc[:,i] = ha_acclim(Tv,Kc25,kcha_range[i])

for i in range(10):
    plt.plot(Tv,Kc[:,i],color = cmap(i))
plt.title('kcha (+-50%)')
plt.ylabel('Kc')
plt.xlabel('Tveg (K)')
```

```python
# Ko
Tv = np.arange(285,313,1)
Ko = np.empty((len(Tv),10))
for i in range(10):
    Ko[:,i] = ha_acclim(Tv,Ko25,koha_range[i])

for i in range(10):
    plt.plot(Tv,Ko[:,i],color = cmap(i))
plt.title('koha (+-50%)')
plt.ylabel('Ko')
plt.xlabel('Tveg (K)')
```

```python
# cp
Tv = np.arange(285,313,1)
cp = np.empty((len(Tv),10))
for i in range(10):
    cp[:,i] = ha_acclim(Tv,cp25,cpha_range[i])

for i in range(10):
    plt.plot(Tv,cp[:,i],color = cmap(i))
plt.title('cpha (+-50%)')
plt.ylabel('cp')
plt.xlabel('Tveg (K)')
```
