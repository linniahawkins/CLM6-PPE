### Plant trait coordination

```python
# Define triangle vertices
vertices = np.array([[0, 0], [1, 0], [0.5, np.sqrt(3) / 2]])

# Function to generate random points within the triangle
def random_point_in_triangle(v1, v2, v3, n=100):
    r1 = np.sqrt(np.random.rand(n))
    r2 = np.random.rand(n)
    points = (1 - r1)[:, None] * v1 + (r1 * (1 - r2))[:, None] * v2 + (r1 * r2)[:, None] * v3
    return points

# Generate random points
points = random_point_in_triangle(vertices[0], vertices[1], vertices[2], n=100)

# Plot the triangle and scatter plot
plt.figure(figsize=(6, 6))
plt.plot(*zip(*np.append(vertices, [vertices[0]], axis=0)), 'k-')  # Draw triangle
plt.scatter(points[:, 0], points[:, 1], color="blue", s=10)         # Scatter points
plt.xlim(-0.1, 1.1)
plt.ylim(-0.1, 1)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Scatter Plot on a Triangle")
plt.show()
```

```python
# Load the NetCDF file
in_dir = '/glade/derecho/scratch/linnia/ctsm5.3.0_transient_AD_oaat0000/run/'
ds = xr.open_dataset(in_dir + "ctsm5.3.0_transient_AD_oaat0000.clm2.h0.0021-01-01-00000.nc")

# Initialize an empty DataFrame to hold all variables
df_all = pd.DataFrame()

# Loop through each variable in the dataset
for var_name in ds.variables:
    # Flatten the variable data to 1D and convert it to a DataFrame
    var_data = ds[var_name].values.flatten()
    var_df = pd.DataFrame({var_name: var_data})
    # Concatenate to the main DataFrame
    df_all = pd.concat([df_all, var_df], axis=1)
```
